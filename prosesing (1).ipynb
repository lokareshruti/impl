```python
import numpy as np
import pandas as pd

```

```python
#import
ds=pd.read_csv("data.csv")
ds
```

```python
# encoding
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encode_data=ohe.fit_transform(ds[['Country']])
encode_data
```

```python
#Creating columes
column_names=ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encode_data,columns= column_names)
encoded_df
```

```python
#joingoriginal dataset and newdatafreame
ds=pd.concat([ds,encoded_df],axis=1)
ds
```

```python
#Lable Encoding

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
ds
```

```python
#features scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ds[['Age','Salary']]=sc.fit_transform
```

```python
ds
```

```python
#divided dataset into dependent and independent

x=ds[['Age','Salary','Purchased','Country_France','Country_Germany','Country_Spain']]
y=ds[['Purchased']]
```

```python
y
```

```python
#spliting dataset as training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```
